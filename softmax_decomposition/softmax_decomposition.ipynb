```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.arange(-5, 5, 0.1)
y = np.exp(x)
fig, ax = plt.subplots()
ax.set_aspect("equal", adjustable="box")
plt.ylim(0, 10)
plt.xlim(-4, 4)

ax.spines["left"].set_position("zero")
ax.spines["bottom"].set_position("zero")
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")

ax.set_yticks(ax.get_yticks()[1:])

ax.plot(x, y)
ax.text(2, 4, r"$y=e^x$")
```

```python
from scipy.special import softmax

values = [-0.3124, 1.4719, 2.1271, -1.0342]
exp_values = np.around(np.exp(values), decimals=4)
softmax_values = np.around(softmax(values), decimals=4)
aggregated_values = np.concatenate((values, exp_values, softmax_values))
margin = 0
x_min, x_max = np.min(aggregated_values) - margin, np.max(aggregated_values) + margin
colors = ["#4A55A2", "#7895CB", "#A0BFE0", "#C5DFF8"]
y_pos = np.arange(len(values))
width = 1

fig, [ax1, ax2, ax3, ax4, ax5, ax6, ax7] = plt.subplots(
    7, 1, gridspec_kw={"height_ratios": [4, 2, 4, 2, 4, 1, 1]}
)
fig.subplots_adjust(hspace=1)
fig.set_figwidth(25)
fig.set_figheight(8)

ax1.set_title("Some random values", fontweight="bold")
ax1.barh(y_pos, values, width, color=colors)
ax1.set_yticks(y_pos, labels=values)
ax1.set_xlim(x_min, x_max)
ax1.set_xticks(np.arange(int(x_min), x_max, 1))
ax1.spines[["right", "top"]].set_visible(False)


ax2.annotate(
    "",
    xy=(0.5, 0),
    xytext=(0.5, 1),
    arrowprops=dict(facecolor="black", headwidth=6, headlength=6, width=2),
)
ax2.text(
    0.58,
    0.5,
    "exponential function (1)",
    horizontalalignment="center",
    verticalalignment="center",
    transform=ax2.transAxes,
    fontsize=15,
    backgroundcolor="white",
    style="italic",
)
ax2.axis("off")

ax3.set_title("Positive values", fontweight="bold")
ax3.barh(y_pos, exp_values, width, color=colors)
ax3.set_yticks(y_pos, labels=exp_values)
ax3.set_xlim(x_min, x_max)
ax3.set_xticks(np.arange(0, x_max, 1))
ax3.spines[["right", "top"]].set_visible(False)

ax4.annotate(
    "",
    xy=(0.5, 0),
    xytext=(0.5, 1),
    arrowprops=dict(facecolor="black", headwidth=6, headlength=6, width=2),
)
ax4.text(
    0.56,
    0.5,
    "normalization (2)",
    horizontalalignment="center",
    verticalalignment="center",
    transform=ax4.transAxes,
    fontsize=15,
    backgroundcolor="white",
    style="italic",
)
ax4.axis("off")

ax5.set_title("Probabilities", fontweight="bold")
ax5.barh(y_pos, softmax_values, width, color=colors)
ax5.set_yticks(y_pos, labels=softmax_values)
ax5.set_xlim(x_min, x_max)
ax5.set_xticks([0, 1])
ax5.axvspan(0, 1, color="b", alpha=0.03)
ax5.spines[["right", "top"]].set_visible(False)

ax6.axis("off")

ax7.set_title("Probabilities add to 1", fontweight="bold")
for i, v in list(enumerate(softmax_values)):
    ax7.barh(0, v, width, left=sum(softmax_values[:i]), color=colors[i])
ax7.set_yticks([0], labels=[1])
ax7.set_xlim(x_min, x_max)
ax7.set_xticks([0, 1])
ax7.spines[["right", "top"]].set_visible(False)

plt.savefig("all.png", dpi=600)
```
